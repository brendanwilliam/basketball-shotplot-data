# Building shot gravity chart from the Celtics' 2023-2024 season

```python
import os
import sys
import pandas as pd

fp = '../../data/processed/S2324/team/BOS_games.csv'
df = pd.read_csv(fp)
```

```python
df.head()
```

```python
df = df[(df['teamTricode'] == 'BOS') & (df['isFieldGoal'] == 1)]
df.sample(10)
```

```python
cols = ['xLegacy', 'yLegacy', 'scoreVal']
df = df[cols].reset_index(drop=True)
df.sample(10)
```

```python
df.shape
```

```python
df['m'] = df['scoreVal'] - 1
df.head()
```

```python
df = df.rename(columns={'xLegacy': 'x', 'yLegacy': 'y'})
df = df.drop(columns=['scoreVal'])
df.head()
```

```python
df.shape
```

```python
df.to_csv('BOS_shotgravity.csv', index=False)
```
