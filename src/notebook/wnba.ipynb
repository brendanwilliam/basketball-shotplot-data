```python
# Imports
from bs4 import BeautifulSoup
import os
import requests
import json
import re
import time
import random
```

```python
# Constants

# URL Constants
URL_START = "https://stats.wnba.com/game/" # First part of the URL
URL_END = "/shotchart"  # Last part of the URL

# WNBA Constants
NUM_TEAMS = 12  # Number of teams in the WNBA
NUM_GAMES = 40  # Number of games each team plays in a season
TOTAL_GAMES = (NUM_TEAMS / 2) * NUM_GAMES # Total number of games in a season

# URL List for all games in the 2024 season
URL_LIST = []
URL_GAME_ID = 1022400001  # First game of the 2024 season

for i in range(int(TOTAL_GAMES)):
    game_id = str(URL_GAME_ID + i)
    URL_LIST.append(URL_START + game_id + URL_END)

# Request headers
HEADERS = {
  'Access-Control-Allow-Origin': '*',
  'Access-Control-Allow-Methods': 'GET',
  'Access-Control-Allow-Headers': 'Content-Type',
  'Access-Control-Max-Age': '6000',
  'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'
}

# Export directory
EXPORT_PATH = "../src/data/raw/"
```

```python
# Functions

def get_soup(url):
    """
    Args:
        url (str): URL to scrape
    Returns:
        soup (BeautifulSoup): BeautifulSoup object of the URL
    """
    try:
        req = requests.get(url, headers=HEADERS)
        print(req.text)
        return BeautifulSoup(req.text, "lxml")
    except:
        print("Error: Unable to retrieve information from the given URL:\n\t" + url)

soup = get_soup(URL_LIST[0])
print(soup)
```

```python
def save_soup(soup, filename):
    """
    Args:
        soup (BeautifulSoup): BeautifulSoup object to save
        filename (str): Filename to save the BeautifulSoup object
    Returns:
        None
    """
    with open(filename, "w") as file:
        file.write(str(soup))

save_soup(soup, "../data/raw/test.html")
```
